```python
class Tabuada:
    def __init__(self, numero: int):
        self.numero = numero

    def validar_numero(self) -> bool:
        return 1 <= self.numero <= 10

    def gerar(self) -> list:
        if not self.validar_numero():
            raise ValueError("O número deve estar entre 1 e 10.")
        return [f"{self.numero} x {i} = {self.numero * i}" for i in range(1, 11)]


def solicitar_numero():
    while True:
        try:
            numero = int(input("Digite um número entre 1 e 10: "))
            if 1 <= numero <= 10:
                return numero
            else:
                print("O número deve estar entre 1 e 10.")
        except ValueError:
            print("Entrada inválida. Digite um número inteiro.")


numero = solicitar_numero()
tabuada = Tabuada(numero)

print("\nTabuada:\n")
for linha in tabuada.gerar():
    print(linha)
```
